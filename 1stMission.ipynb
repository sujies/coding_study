````python
import pandas as pd
df = pd.read_csv('/content/서울특별시 강남구_생활폐기물배출량_20221019.csv', encoding="cp949")
df.head()
````

````python
#  2019년과 2020년의 생활 폐기물 차이
df_new = df.copy()
df_new['gap'] = df_new.iloc[:,1] -  df_new.iloc[:,-1]
df_new.head()
````

````python
# 증가 감소 표시하기
def comparison_gap(x):
    if x >=0:
        return '증가'
    else:
        return '감소'

df_new['비교'] = df_new['gap'].apply(comparison_gap)
df_new.head()
````
